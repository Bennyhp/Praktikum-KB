### Import library
> Library yang digunakan antara lain adalah pandas,matplotlib dan seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Membaca File CSV
> Membaca file csv menggunakan pandas

```python
df =  pd.read_csv('songs_normalize.csv')
```

### Grafik Pie
> ##### Kesimpulan Visualisasi :
> * Pada grafik pie dibawah menunjukkan bahwa 27,5% dari 2000 lagu didalam dataset mengandung         kata-kata _explicit_
> * Dan grafik juga menunjukkan bahwa 72,5% dari 2000 lagu didalam dataset tidak mengandung kata-kata _explicit_


```python
data_clean = df['explicit']
explicity = df['explicit']

data_graph = data_clean.value_counts().rename_axis('key').reset_index(name='id_count')

plt.pie(data_graph.id_count, labels=data_graph.key, startangle=90, autopct='%.1f%%')
plt.legend(bbox_to_anchor=(1, 0.2))
plt.title('Tingkat musik yang mengandung kata-kata explicit')
plt.xlabel('Explicit')
plt.ylabel('Not Explicit')
plt.show()
```

### Grafik Bar
> #### Kesimpulan Visualisasi
> * Pada grafik dibawah, men-visualisasikan tingkat _danceability_ tiap tahunnya, semakin mendekati 1.0 nilainya maka semakin tinggi skor _danceability_-nya

```python
x = df['year']
y = df['danceability']

fg = plt.figure()
fg.set_figwidth(10)
fg.set_figheight(5)
plt.bar(x, y)
plt.title('Tingkat danceability lagu tiap tahunnya')
plt.xlabel('Tahun')
plt.ylabel('Danceability')
plt.show()
```

### Grafik Scatterplot
> #### Kesimpulan Visualisasi :
> * Pada grafik line dibawah menunjukkan tingkat _energy_ dan _danceability_ yang dimana banyak terkumpul di range _energy_ 0.6-0.8 dan range _danceability_ 0.6-0.8 
> * Jadi dapat disimpulkan bahwa semakin tinggi nilai _danceability_ maka kemungkinan _energy_ untuk pendengar skornya semakin tinggi pula, namun dapat juga dilihat ada beberapa titik yg _danceability_nya tinggi namun _energy_nya rendah, ini juga dapat terpengaruh dari _genre_ lagu itu sendiri

```python
x = df['energy']
y = df['danceability']
plt.xlabel('energy')
plt.ylabel('Danceability')
plt.scatter(x,y)
plt.show()
```
